### Распределение данных о клиентах банка

Изучим, как распределяются клиенты банка по возрасту, уровню зарплаты, балансу на счете и баллам кредтиного скоринга. 

#### Возраст клиентов

Посмотрим на то, как клиенты распределены по возрасту.

```python
# Распределение по возрасту

plt.figure(figsize = (15,5))

ax1 = plt.subplot(1,2,1)
(
    sns.distplot(data['age'], bins = 75),
    plt.title('Распределение клиентов по возрасту'),
    plt.xlabel("Возраст"),
    plt.ylabel("Количество клиентов")
)

ax2 = plt.subplot(1,2,2)
(
    sns.boxplot(x=data['age'], palette="Blues"),
    plt.title('Распределение'),
    plt.xlabel("Возраст"),
    plt.show()
); 
```

Распределение клиентов по возрасту похоже на нормальное. В основном это люди в возрасте 32-44 лет, есть несколько пожилых клиентов. 

Посмотрим, как распределены по возрасту клиенты, которые ушли и остались.

```python
plt.figure(figsize = (12,5))
sns.distplot(data_bank['age'])
sns.distplot(data_churn['age'])
plt.title('Распределение клиентов по возрасту')
plt.ylabel('Количество клиентов')
plt.xlabel('Возраст')
plt.legend(['Остались', 'Ушли'])
plt.show()
```

Отчетливо виден тренд на то, что уходят клиенты более зрелого возраста. Это примерно группа 45+, возможно, что им не нравятся продукты банка, ориентированные на зрелых людей. Речь может идти о накопительных и страховых пенсионных продуктах.

#### Уровень зарплаты

Изучим распределение по уровню зарплаты.

```python
# Распределение по уровню зарплаты

plt.figure(figsize = (15,5))

ax1 = plt.subplot(1,2,1)
(
    sns.distplot(data['estimated_salary'], bins = 50),
    plt.title('Распределение клиентов по уровню зарплаты'),
    plt.xlabel("Уровень зарплаты"),
    plt.ylabel("Количество клиентов"),
    plt.xticks(rotation=45)
)

ax2 = plt.subplot(1,2,2)
(
    sns.boxplot(x=data['estimated_salary'], palette="Blues"),
    plt.title('Распределение'),
    plt.xlabel("Уровень зарплаты"),
    plt.xticks(rotation=45),
    plt.show(),
); 
```

Клиенты банка по уровню зарплаты распределены равномерно. Посмотрим, верно ли это в разрезе ушедших клиентов. 

```python
plt.figure(figsize = (12,5))
sns.distplot(data_bank['estimated_salary'])
sns.distplot(data_churn['estimated_salary'])
plt.title('Распределение клиентов по уровню зарплаты')
plt.ylabel('Количество клиентов')
plt.xlabel('Зарплата')
plt.legend(['Остались', 'Ушли'])
plt.show()
```

Значительных отличий в уровне зарплаты между оставшимися и ушедшими клиентами не замечено.

#### Баланс на счете

Посмотрим, как распределяются клиенты по уровню баланса на счете.

```python
# Распределение баланса на счете

plt.figure(figsize = (15,5))

ax1 = plt.subplot(1,2,1)
(
    sns.distplot(data['balance'], bins = 50),
    plt.title('Распределение клиентов по уровню баланса'),
    plt.xlabel("Баланс"),
    plt.ylabel("Количество клиентов"),
    plt.xticks(rotation=45)
)

ax2 = plt.subplot(1,2,2)
(
    sns.boxplot(x=data['balance'], palette="Blues"),
    plt.title('Распределение'),
    plt.xlabel("Баланс"),
    plt.xticks(rotation=45),
    plt.show(),
); 
```

Распределение данных выглядит нормально. Посмотрим, есть ли связь между балансом на счете и оттоком клиентов.

## Проверка гипотез

Сформулируем и проверим статистические гипотезы.

Гипотеза 1:
- H0: Доходы клиентов, которые ушли и которые остались не отличаются.
- H1: Доходы клиентов, которые ушли и которые остались отличаются.

Гипотеза 2:
- H0: Количество продуктов у клиентов, которые ушли и которые остались не отличается.
- H1: Количество продуктов у клиентов, которые ушли и которые остались разное.

### Гипотеза 1

Проверим первую гипотезу.

- H0: Доходы клиентов, которые ушли и которые остались не отличаются.
- H1: Доходы клиентов, которые ушли и которые остались отличаются.

Данные о доходе (зарплате) клиентов распределены равномерно, а не нормально, поэтому мы не можем применить t-критерий Стьюдента. Для проверки применим критерий Манна-Уитни, он подходит для сравнения распределения значений двух независимых выборок.

Уровень статистической значимости - 5%.

```python
alpha = 0.05

results = st.mannwhitneyu(
    data_bank['estimated_salary'], 
    data_churn['estimated_salary']
)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод**: Доходы клиентов, которые ушли и которые остались не отличаются. Нулевую гипотезу отвергнуть не можем.

### Гипотеза 2

- H0: Количество продуктов у клиентов, которые ушли и которые остались не отличается.
- H1: Количество продуктов у клиентов, которые ушли и которые остались разное.

Данные о продуктах также распределены не нормально, поэтому применяем критерий Манна-Уитни.

Уровень статистической значимости - 5%.

```python
alpha = 0.05

results = st.mannwhitneyu(
    data_bank['products'], 
    data_churn['products']
)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод**: Количество продуктов у клиентов, которые ушли и которые остались разное, отвергаем нулевую гипотезу. Действительно, как мы показали ранее, клиенты с большим числом продуктов склонны уходить чаще. 

### Вывод
