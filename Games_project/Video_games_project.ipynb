# Исследование о продажах компьютерных игр

Исследование направлено на изучение рынка компьютерных игр по данным из открытых источников. Цель - исследовать исторические данные о продажах компьютерных игр для разных платформ, чтобы использовать их в будущем для прогнозирования продаж.

Мы выполним предобработку данных и изучим их, чтобы выявить существующие закономерности на исследуемом рынке. Также посмотрим на пропуски, дубликаты и причины их возникновения.

В ходе исследовательского анализа мы посмотрим, сколько игр выпускалось в разные годы, как менялись продажи по платформам, рассмотрим, какие из них являются потенциально прибыльными. Также мы посмотрим, какие жанры представляются наиболее прибыльными. Кроме того, мы изучим, как на продажи влияют отзывы критиков и пользователей, и составим портрет пользователя по регионам мира.

Откроем датафрейм и изучим общую информацию

Однако есть ряд моментов, которые необходимо исправить для последующего анализа:
- привести заголовки столбцов к нижнему регистру
- исправить тип данных в колонках с годом релиза, оценкой пользователей

## Предобработка данных

### Приведение к нижнему регистру
Приведем названия столбцов к нижнему регистру

```python
data.columns = data.columns.str.lower() 
data.head()
```

### Преобразование данных в нужные типы

Преобразуем данные о годе выпуска из float в int. Подставлять другие значения не стоит, поскольку это может исказить данные при анализе.

```python
data = data.dropna(subset=['year_of_release'])
data['year_of_release'] = data['year_of_release'].astype('int')
data.head()
```

Преобразуем данные о пользовательской оценке в тип float. Они нужны нам именно как числа для последующего анализа.

```python
data['user_score'].unique()
```

Нам мешает это сделать значение 'tbd'. Обычно оно означает 'to be decided'. То есть оценка не выставлена пользователем по какой-то причине. Заменим на nan и приведем к типу float.

```python
data.loc[data['user_score'] == 'tbd','user_score'] = np.nan
data['user_score'] = data['user_score'].astype('float')
data.info()
```

### Обработка пропусков

Посмотрим на пропуске в таблице

```python
data.isna().sum()
```

Мы видим по два пропуска в столбцах названия игры и жанра. Значение небольшое, можно их удалить.
Основная часть пропущенных значений приходится на столбцы с оценками критиков, пользователей и рейтингом. Эти данные важны для анализа, поэтому их необходимо заполнить. 

Мы не можем подставить ни среднее, ни медиану, поскольку не знаем, какую оценку поставил бы критик или пользователь. Наши домыслы могут исказить данные и повлиять на прогнозирование продаж. Поставим им рейтинг -1. В последнем столбце поставим unknown.

К пропускам в данных могли привести нежелание пользователя или критика выставлять оценку игре. Возможно, что оценка не подгрузилась из какой-то системы или при переносе данных. Можно предположить также, что на каких-то платформах в принципе отсутствовала возможность поставить оценку. 

Графа "рейтинг" может быть пустой из-за отсутствия актуальных данных от компании Entertainment Software Rating Board (ESRB).

Удалим данные с пропусками в названии и жанре.

```python
data = data.dropna(subset=['name', 'genre'])
data.isna().sum()
```

```python
data['critic_score'] = data['critic_score'].fillna(-1)
data['user_score'] = data['user_score'].fillna(-1)
data['rating'] = data['rating'].fillna('unknown')
data.head()
```

```python
data.duplicated().sum() #Проверим на всякий случай дубликаты
```

### Добавление столбца с продажами во всех регионах
Подсчитаем суммарные продажи во всех регионах и запишем в отдельный столбец.

```python
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
data.head()
```

### Выводы

Мы изучили данные и провели предобработку. Удалили строки с пропусками в названии игры и жанре, а также с отсутствующим годом выпуска. Заменили тип данных для столбца "пользовательская оценка".

Обнаружили, что значительная часть пропусков в данных приходится на столбцы с оценками критиков, пользователей и рейтингом. Эти данные важны для анализа. Мы не можем подставить ни среднее, ни медиану. Наши домыслы могут исказить данные и повлиять на прогнозирование продаж. Поставили им рейтинг -1. 

Пропуски могли возникнуть, если пользователь или критик на захотел выставлять оценку игре. Возможно, что оценка не подгрузилась из какой-то системы или при переносе данных. Также возможно, что на каких-то платформах в принципе отсутствовала возможность поставить оценку.

Подсчитали суммарные продажи по всем регионам для каждой строки и вынесли их в отдельный столбец.

### Запуск игр по платформам

Рассмотрим, как менялись продажи по отдельным платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам.

Можно сказать, что наиболее активные продажи пришлись на период с 2000 по 2006 год, когда вышло новое поколение PS.

Приставка XBox360 появилась в 2005 году. Динамика продаж была неравномерной и достигла пика в 2013 году, затем продажи стали снижаться, когда появилось следующее поколение - XBox One (в ноябре 2013 года).

```python
#PS3

(
    data[(data['platform'] == 'PS3')]
    .pivot_table(index='year_of_release', values='total_sales')
    .plot(grid=True, figsize=(12, 5))
); 
```

В 2006 году появилось новое поколение PS - PS3, график показывает, что продажи приставки были стабильно высокими в течение шести лет и достигли пика в 2013 году. Затем продажи стали снижаться. Действительно, в ноябре 2013 года вышла PS4.

Изучив данные топ-3 популярных игровых консолей, можно сказать, что "срок жизни" игровой платформы составляет порядка десяти лет, рост продаж и их пик приходится на первые три-четыре года жизни, всего "активная жизнь" длится шесть-восемь лет.

### Данные для прогноза

Для прогнозирования продаж на 2017 год возьмем за основу данные 2013-2016 годов. Это период позволит нам увидеть, продажи каких платформ растут, находятся на пике или снижаются, и выявить те, которые в настоящий момент находятся на этапе роста продаж. В этом периоде представлены данные по всем актуальным консолям. Также мы сможем увидеть, в каком году появились новые платформы, если они были.

Мы также включаем данные за 2016 год, потому что исследование проводится в декабре, основные данные по году уже присутствуют. Возможен некоторый пик продаж перед Новым годом, но вряд ли он сильно изменит картину по году. 

```python
forecast_data = data.query('year_of_release >= 2013').reset_index(drop=True)
forecast_data.head()
```

### Платформы-лидеры и продажи

Посмотрим, какие платформы лидируют по суммарным продажам за 2005-2016 годы.

```python
(
    forecast_data
    .groupby('platform')
    .agg({'total_sales':'sum'})
    .sort_values(by='total_sales', ascending=False)
)
```

В исследуемый период по продажам лидируют пять платформ: PS4, PS3, XOne, 3DS и X360. Остальные платформы следуют за ними. Можно предположить, что они на момент исследования уже были в продаже и постепенно теряли актуальность.

Посмотрим, какая динамика продаж была у топ-платформ по годам.

```python
(
    forecast_data
    .pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')
    .plot(grid=True, figsize=(12,5), title="Продажи платформ по годам")
);
```

Действительно, видим, что топ-5 исследуемых платформ за исследуемый период завершали свой жизненный цикл, поэтому они и оказались лидерами по продажам. 

Видим, что продажи игр для платформ WiiU, PC, PSV также снижались за исследуемый период. Действительно, им оставалось недолго: поддержка WiiU прекратилась в 2017 году, а PSV - в 2019 году.

**Вывод:** При прогнозе продаж стоит делать ставку актуальные на данный момент платформы: это персональные компьютеры, а также восьмое поколение игровых консолей. К ним относятся PS4 и XOne. Девятого поколения пока нет на горизонте.

Осенью 2016 года Sony выпустила две новые аппаратные версии приставки: PlayStation 4 Slim, обладающую уменьшенными габаритами, и PlayStation 4 Pro, оснащённую более мощными основным и графическим процессором с расчётом на возможность вывода изображения и потокового видео в разрешении 4K. XBox в августе 2016 года также выпустила новую версию своей консоли - XOne S, а X360, был снят с производства.

Портативные устройства теряют актуальность на фоне появления более мощных смартфонов и планшетов.

### Глобальные продажи по платформам

Построим график «ящик с усами» по глобальным продажам игр в разбивке по платформам.

```python
forecast_data.describe()
```

```python
plt.figure(figsize=(17, 11))
sns.boxplot(data=forecast_data, y='platform', x='total_sales')
plt.xlim(-0.1, 5)

plt.title('Объемы продаж игр на разных платформах', size=17)
plt.xlabel('Объем продаж', size=12)
plt.ylabel('Название платформы', size=12)

plt.show()
```

График показывает, что основная часть продаж для большинства платформ (расстояние между Q1 и Q3) лежит в пределах 0,5 миллиона проданных копий игр в год. Значительно превышают это значение следующие платформы: PS3, PS4, X360, XOne.

### Влияние отзывов

Возьмем для примера платформу PS4, по ней много данных, которые свидетельствуют о ее популярности. Посмотрим, как как на ее продажи влияют отзывы пользователей и критиков. 

Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. 

```python
#Посмотрим на отзывы критиков на игры для PS4

(
    forecast_data.query('platform == "PS4" and critic_score > 0') 
    #Убираем значения -1, которыми заполняли пустоту
    .plot(x='critic_score', y='total_sales', kind='scatter', alpha=0.5)
);
```

Судя по графику, пользователи действительно читают отзывы критиков: чем выше оценка, тем выше показатель продаж.

```python
#Посмотрим на отзывы пользователей на игры для PS4

(
    forecast_data.query('platform == "PS4" and critic_score > 0') 
    .plot(x='user_score', y='total_sales', kind='scatter', alpha=0.5)
);
```

На этом графике видим в целом похожую картину, чем выше оценка пользователей, тем лучше игра продается. В то же время можно отметить, что пользователи настроены более скептически, чем критики - оценка выше 8,5 почти не встречается. Кроме того, есть высокие показатели продаж у игр с оценкой в 4 балла. ### Влияние рейтинга ESRB на продажи

Посмотрим, влияет ли рейтинг ESRB  на продажи в отдельных регионах.

```python
# Определим, какие рейтинги встречаются в датасете

forecast_data['rating'].unique()
```

Создадим столбец с пояснением, что значит каждый рейтинг.

- `AO` - Adults Only 18+ — Только для взрослых
- `E` - Everyone - для всех
- `E10+` - Everyone 10 and older — Для всех от 10 лет и старше
- `EC` - Early childhood — Для детей младшего возраста (от 3 лет)
- `M` - Mature - Для взрослых (старше 17 лет)
- `RP` - Rating Pending — Рейтинг ожидается (ожидается оценка ESRB)
- `T` - Teen — Подросткам (от 13 лет)
- `unknown` - нет данных

```python
def categorize_rating(rating):
    try:
        if rating == 'E':
            return 'everyone'
        elif rating == 'AO':
            return '18+'
        elif rating == 'E10+':
            return '10+'
        elif rating == 'EC':
            return '3+'
        elif rating == 'M':
            return '17+'
        elif rating == 'T':
            return '13+'
        elif rating == 'RP':
            return 'NR' #No Rating
        elif rating == 'unknown':
            return 'NR'
    except:
        pass
```

```python
forecast_data['rating_category'] = forecast_data['rating'].apply(categorize_rating)

forecast_data.head()
```

```python
# Рейтинг и продажи в Северной Америке

(
    forecast_data
    .pivot_table(index='rating_category', values='na_sales')
    .sort_values(by='na_sales', ascending=False)
    .plot(kind='bar', figsize=(7, 5), title="ESRB в Северной Америке")
); 

# Рейтинг и продажи в Европе

(
    forecast_data
    .pivot_table(index='rating_category', values='eu_sales')
    .sort_values(by='eu_sales', ascending=False)
    .plot(kind='bar', figsize=(7, 5), title="ESRB в Европе")
);

# Рейтинг и продажи в Японии

(
    forecast_data
    .pivot_table(index='rating_category', values='jp_sales')
    .sort_values(by='jp_sales', ascending=False)
    .plot(kind='bar', figsize=(7, 5), title="ESRB в Японии")
);
```

## Проверка гипотез

Проверим несколько гипотез.

### Гипотеза I

Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Альтернативная гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC не одинаковые.

Нулевая гипотеза всегда формулируется, чтобы использовать знак равенства, то есть мы проверяем, лежат ли данные в районе изначально предполагаемого значения. Это гипотеза о равенстве средних двух генеральных совокупностей. Проверим ее.

```python
XOne_stat = forecast_data[(forecast_data['platform'] == 'XOne') & (forecast_data['user_score'] > 0)]['user_score']
PC_stat = forecast_data[(forecast_data['platform'] == 'PC') & (forecast_data['user_score'] > 0)]['user_score']
```

```python
alpha = 0.05

results = st.ttest_ind(
    XOne_stat, 
    PC_stat)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

P-value оказался выше критического уровня статистической значимости, поэтому нулевую гипотезу принимаем. **Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.**

### Гипотеза II

Нулевая гипотеза: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые.

Альтернативная гипотеза: Средние пользовательские рейтинги жанров Action и Sports разные.

Это тоже гипотеза о равенстве средних двух генеральных совокупностей. Проверим ее.

```python
action_stat = forecast_data[(forecast_data['genre'] == 'Action') & (forecast_data['user_score'] > 0)]['user_score']
sports_stat = forecast_data[(forecast_data['genre'] == 'Sports') & (forecast_data['user_score'] > 0)]['user_score']
```

```python
alpha = 0.05

results = st.ttest_ind(
    action_stat, 
    sports_stat)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
    
# Здесь я подозревала ошибку в коде
```
